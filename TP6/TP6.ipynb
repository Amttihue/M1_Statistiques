```python
a = ss.norm.ppf(0.025)
b = ss.norm.ppf(0.975)
print(f"L'intervalle centré est [{a:.3f}, {b:.3f}]")
print(f"On a bien ss.norm.cdf(a) (= {ss.norm.cdf(a):.3f}) = 0.025)")
```

On effectue les 4 étapes pour définir notre test sur la variance

1) On définit $H_0$ :
   L'échantillon n°2 provient de la population de l'échantillon n°1
2) On considère la V.A. $F=\frac{(n_1-1) {s'_1}^2}{(n_2-1) {s'_2}^2}$ qui suit une loi Fisher-Snedecor de degrés de libertés $(n_1-1;n_2-1)$ (in $(199,249)$)
